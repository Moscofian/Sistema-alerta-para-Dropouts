### Descrição das Colunas:

1. **School (Escola):** Nome da escola frequentada (ex: MS).
2. **Gender (Gênero):** Gênero do estudante (ex: M para Masculino, F para Feminino).
3. **Age (Idade):** Idade do estudante.
4. **Address (Endereço):** Tipo de residência (U para urbana, R para rural).
5. **Family_Size (Tamanho_da_Família):** Tamanho da família (GT3 para maior que 3 pessoas, LE3 para menor ou igual a 3 pessoas).
6. **Parental_Status (Estado_dos_Pais):** Situação de coabitação dos pais (A para morando juntos, T para morando separados).
7. **Mother_Education (Escolaridade_da_Mãe):** Nível de escolaridade da mãe (0 a 4).
8. **Father_Education (Escolaridade_do_Pai):** Nível de escolaridade do pai (0 a 4).
9. **Mother_Job (Profissão_da_Mãe):** Tipo de trabalho exercido pela mãe.
10. **Father_Job (Profissão_do_Pai):** Tipo de trabalho exercido pelo pai.
11. **Reason_for_Choosing_School (Motivo_da_Escolha):** Razão pela qual escolheu a escola (ex: curso).
12. **Guardian (Responsável):** Responsável legal pelo estudante (ex: mãe).
13. **Travel_Time (Tempo_de_Viagem):** Tempo de deslocamento até a escola (em minutos).
14. **Study_Time (Tempo_de_Estudo):** Horas de estudo semanais (escala de 1 a 4).
15. **Number_of_Failures (Número_ de_Reprovações):** Quantidade de reprovações em turmas anteriores.
16. **School_Support (Apoio_Escolar):** Se o estudante recebe apoio educacional extra da escola (sim/não).
17. **Family_Support (Apoio_Familiar):** Se a família oferece apoio educacional em casa (sim/não).
18. **Extra_Paid_Class (Aulas_Particulares):** Participação em aulas extras pagas/reforço (sim/não).
19. **Extra_Curricular_Activities (Atividades_Extracurriculares):** Envolvimento em atividades extracurriculares (sim/não).
20. **Attended_Nursery (Frequentou_Creche):** Se frequentou a creche/educação infantil quando criança (sim/não).
21. **Wants_Higher_Education (Deseja_Ensino_Superior):** Desejo de cursar o ensino superior/faculdade (sim/não).
22. **Internet_Access (Acesso_à_Internet):** Disponibilidade de internet em casa (sim/não).
23. **In_Relationship (Namorando):** Status de relacionamento amoroso (sim/não).
24. **Family_Relationship (Relacionamento_Familiar):** Qualidade das relações familiares (escala de 1 a 5).
25. **Free_Time (Tempo_Livre):** Quantidade de tempo livre após a escola (escala de 1 a 5).
26. **Going_Out (Sair_com_Amigos):** Frequência de saídas com amigos (escala de 1 a 5).
27. **Weekend_Alcohol_Consumption (Consumo_Álcool_Fim_de_Semana):** Consumo de álcool nos fins de semana (escala de 1 a 5).
28. **Weekday_Alcohol_Consumption (Consumo_Álcool_Dias_Úteis):** Consumo de álcool em dias úteis (escala de 1 a 5).
29. **Health_Status (Estado_de_Saúde):** Avaliação do estado de saúde do estudante (escala de 1 a 5).
30. **Number_of_Absences (Número_de_Faltas):** Quantidade total de faltas na escola.
31. **Grade_1 (Nota_1):** Nota recebida na primeira avaliação (geralmente primeiro bimestre/trimestre).
32. **Grade_2 (Nota_2):** Nota recebida na segunda avaliação.
33. **Final_Grade (Nota_Final):** Nota final recebida no fim do ano (G3).
34. **Dropped_Out (Evadiu):** Indicador se o aluno abandonou a escola ou não (True para Sim / False para Não).

# Pré-Processamento e Machine Learning

```python
# Imports
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import joblib
```

```python
# Transforma os valores String em números
colunas_cat = df.select_dtypes(include=['object', 'category']).columns
encoder=LabelEncoder()

df[colunas_cat]=encoder.fit_transform(colunas_cat)
```

```python
# Strings transformadas em números
df.head()
```

```python
# Tira as colunas para evitar vazamento de dados
# A nota final de alguns alunos é zero por eles sairem no meio do ano letivo
# Por isso foram removidas as colunas 'Final_Grade' e 'Grade_2' para evitar atalhos
X = df.drop(columns=['Dropped_Out', 'Final_Grade', 'Grade_2'])
y = df['Dropped_Out']
```

```python
# Divide os dados
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Pipeline 
# Criação do Pipeline
pipeline = Pipeline([
    
    ('scaler', StandardScaler(with_mean = False)),  # Padroniza os vetores númericos criados
    
    ('logreg', LogisticRegression(solver = 'liblinear', random_state = 42, max_iter = 1000)) 
])
```

```python
# Definir o grid de hiperparâmetros para otimização
parametros_grid = {
    'logreg__class_weight': ['balanced'],   # Balanceia o foco pois a taxa de evasão é baixa
    'logreg__C': [0.1, 1, 10],
    'logreg__penalty': ['l1', 'l2'],
    'logreg__max_iter': [5000, 6000]
}
```

```python
# Configurar o objeto GridSearchCV
grid_search = GridSearchCV(
    pipeline,              # Pipeline com as etapas de pré-processamento e modelo
    parametros_grid,       # Dicionário com as combinações de hiperparâmetros a serem testadas
    cv = 5,                # Número de divisões para validação cruzada (5-fold cross-validation)
    n_jobs = -1,           # Usa todos os núcleos disponíveis do processador para acelerar o processo
    scoring = 'f1',        # Métrica usada para avaliar o desempenho de cada combinação (aqui, acurácia)
    verbose = 1            # Nível de detalhamento do output durante a execução (1 exibe progresso básico)
)
```

```python
# Treinamento com o Pipeline
print("\nIniciando o treinamento do modelo com otimização de hiperparâmetros...\n")
grid_search.fit(X_treino, y_treino)
```

```python
print("\nMelhores hiperparâmetros encontrados:\n")
print(grid_search.best_params_)
```

```python
# Obter o melhor modelo
melhor_modelo = grid_search.best_estimator_
```

# Avaliação do Modelo e Interpretação de Métricas

```python
# Previsões no conjunto de teste
y_pred = melhor_modelo.predict(X_teste)
```

```python
# Relatório completo de métricas
print("\n--- Relatório de Classificação ---")
print(classification_report(y_teste, y_pred))
```

```python
# Visualizar a Matriz de Confusão
ConfusionMatrixDisplay.from_predictions(y_teste, y_pred, cmap='Blues')
plt.title("Matriz de Confusão")
plt.show()
```
